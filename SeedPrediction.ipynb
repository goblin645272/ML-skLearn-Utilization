```python
import pandas as pd
import sklearn
import numpy as nm
from sklearn import svm
from svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
total_data = pd.read_csv('Seed_Data.csv')
```

```python
x = total_data.iloc[:,0:7]
```

```python
y = total_data.iloc[:,7]
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size = 0.2,random_state = 13)
```

```python
sc = StandardScaler()
Xtrain = sc.fit_transform(Xtrain)
Xtest = sc.transform(Xtest)
```

```python
clf = svm.SVC()
clf.fit(Xtrain,Ytrain)
```

```python
pred_clf = clf.predict(Xtest)
```

```python
#accuracy score 
sklearn.metrics.accuracy_score(Ytest,pred_clf)
```

```python
print(sklearn.metrics.classification_report(Ytest,pred_clf))
```
